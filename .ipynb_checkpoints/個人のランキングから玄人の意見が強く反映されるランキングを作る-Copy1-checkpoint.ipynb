```python
import numpy as np
from random import shuffle
```

```python
N = 100 # 作品数
M = 5000 # 5000票
L = N // 2 # 読んだ作品数の上限

index_name_set_dic = {i:str(i)+"th" for i in range(N)} # 50作品生成
t_lst = [i for i in range(N)]
counter_arr = np.zeros((N,N)) + 1
point_arr = np.zeros((N,N))
p1, p2, p3 = 0, 0, 0
for i in range(M):#5000票
    t_real = L*np.random.rand()
    t = int(t_real + 1) #読んだ作品数
    shuffle(t_lst)
    #ここから、投票数に分布を持たせる
    if i < M*9/10 and t_real <= L*16/17 and np.random.randn() < 2:
        for j in range(int(N/2)): # 見る作品をしぼる&見た作品の順位を決める
            for k in range(N-1):
                if t_lst[k] > t_lst[k+1] and np.random.randn() < 0:
                    t_lst[k], t_lst[k+1] = t_lst[k+1], t_lst[k]
#        print("p1")
        p1 += 1
    elif M*9/10 <= i < M*10/10 and t_real <= L*16/17 and np.random.randn() < 2:
        for j in range(int(N/2)): # 見る作品をしぼる&見た作品の順位を決める
            for k in range(N-1):
                if t_lst[k] < t_lst[k+1] and np.random.randn() < 0:
                    t_lst[k], t_lst[k+1] = t_lst[k+1], t_lst[k]
#        print("p2")
        p2 += 1
    else:
        for j in range(int(t/2)): # 見た作品の順位を決める
            for k in range(t):
                if t_lst[k] < t_lst[k+1] and np.random.randn() < 0:
                    t_lst[k], t_lst[k+1] = t_lst[k+1], t_lst[k]
#        print("p3")
        p3 += 1
    #ここまで、投票数に分布を持たせる
    for j in range(t - 1):
        for k in range(j+1, t):
            pibot_t = t_lst[j]
            point_arr[pibot_t, t_lst[k]] += 1
            point_arr[t_lst[k], pibot_t] -= 1
            counter_arr[pibot_t, t_lst[k]] += 1
            counter_arr[t_lst[k], pibot_t] += 1
#    print(t_lst[:t])
point_arr /= counter_arr**(4/5)
point_arr -= np.sum(point_arr) / N**2
```

```python
print(p1, p2, p3)
```

```python
sum_point_arr = np.sum(point_arr, axis=1)
point_lst = []
for i,p in enumerate(sum_point_arr):
    point_lst.append((p,i))
point_lst.sort(reverse=True)
point_lst

ranking_lst = []
for i in range(N):
    ranking_lst.append(point_lst[i][1])
ranking_lst[:5]
```
